```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.groupby(df["Data Venda"].dt.year)["Valor Venda"].sum().plot.bar(title="Vendas x Ano")
plt.xlabel("Ano")
plt.ylabel("Vendas")
plt.xticks(rotation='horizontal');
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Valor Venda"].sum().plot.bar(title="Vendas x Mês (2009)")
plt.xlabel("Mês")
plt.ylabel("Vendas")
plt.xticks(rotation='horizontal');
```

```python
df_2008 = df[df["Data Venda"].dt.year == 2008]
```

```python
df_2008.groupby(df_2008["Data Venda"].dt.month)["Valor Venda"].sum().plot.bar(title="Vendas x Mês (2008)")
plt.xlabel("Mês")
plt.ylabel("Vendas")
plt.xticks(rotation='horizontal');
```

```python
df.to_csv("df_vendas_2008-2009.csv", index=False)
```
